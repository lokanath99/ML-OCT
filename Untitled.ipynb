```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./IRIS.csv')
df.head()
data = df.iloc[:,:-1]
data.head()
```

```python
from sklearn.cluster import KMeans
inir = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=6)
    kmeans.fit(data)
    inir.append(kmeans.inertia_)
    
plt.plot(range(1, 11), inir)
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=6)
kmeans.fit(data)
```

```python
pred = kmeans.predict(data)
cat1 = data[pred == 0]
cat2 = data[pred == 1]
cat3 = data[pred == 2]
plt.scatter(cat1['petal_length'], cat1['petal_width'], c='c')
plt.scatter(cat2['petal_length'], cat2['petal_width'], c='y')
plt.scatter(cat3['petal_length'], cat3['petal_width'], c='r')
plt.show()
```
