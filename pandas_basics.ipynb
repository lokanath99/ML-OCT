```python
import pandas as pd
data = pd.read_csv('./diabetes.csv')
data.tail()
```

```python
# analysing data
data.info()
data.describe()
data.corr()
```

```python
# Data cleaning
data_clean = data.dropna()
data_clean.info()

data.BMI.fillna(data.BMI.mean(), inplace=True)
data.Insulin.fillna(data.Insulin.mode(), inplace=True)
data.info()
```

```python
for i in data.duplicated():
    if i == True:
        print('duplicate present')
        break
    continue
    
data.drop_duplicates(inplace=True)
data.info()
```

```python
data.Pregnancies.fillna(data.Pregnancies.mode(), inplace=True)
data.Glucose.fillna(data.Glucose.mode(), inplace=True)
data.BloodPressure.fillna(data.BloodPressure.mode(), inplace=True)
data.info()
```

```python
data['Test'] = (data.Glucose*data.Insulin)/2
# data.drop('Test', inplace=True)
# data.drop(['Test'], inplace=True)
# data
```

```python
data.drop(columns=['Test', 'SkinThickness'], inplace=True)
data
```
